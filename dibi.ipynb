```python
import pandas as pd
import numpy as np

dibi = pd.read_csv("dibi.csv")
dibi.head()
```

```python
dibi.columns
```

```python
dibi.columns = (
    dibi.columns
    .str.strip()
    .str.replace(" ", "_")
    .str.replace("/", "_")
)
```

```python
# Ubah kolom tanggal DIBI ke datetime
dibi['Tanggal'] = pd.to_datetime(
    dibi['Tanggal___Waktu_Kejadian'],
    errors='coerce'
)

# Buat key tanggal
dibi['Tanggal_Kejadian_Key'] = dibi['Tanggal'].dt.date.astype(str)

# Lihat hasil
dibi[['Tanggal___Waktu_Kejadian', 'Tanggal', 'Tanggal_Kejadian_Key']].head()
```

```python
dibi['kabupaten_lower'] = (
    dibi['Kabupaten']
    .astype(str)
    .str.lower()
    .str.strip()
)
```

```python
tabel_bencana_cuaca = dibi.copy()
```

```python
tabel_bencana_cuaca.to_csv("tabel_bencana_cuaca_clean.csv", index=False)
print("File tersimpan sebagai tabel_bencana_cuaca_clean.csv")
```
